# 🤖 AI Smart Crawler 演示

本Notebook演示如何使用AI Smart Crawler进行智能网页爬取。

```python
# 安装（如果在新的环境中）
# !pip install ai-smart-crawler
```

```python
import asyncio
from ai_smart_crawler import SmartCrawlerSystem

# 初始化
crawler = SmartCrawlerSystem()
print("✅ 爬虫系统初始化完成")
```

## 1️⃣ 快速获取单个网页

```python
# 爬取示例页面
url = "https://httpbin.org/json"
content = await crawler.quick_fetch(url)

print(f"📄 内容预览:")
print(content[:500])
```

## 2️⃣ 批量爬取

```python
urls = [
    "https://httpbin.org/json",
    "https://httpbin.org/html",
]

results = await crawler.batch_crawl(urls)

for url, result in zip(urls, results):
    status = "✅" if result.success else "❌"
    size = len(result.markdown) if result.success else 0
    print(f"{status} {url} ({size} chars)")
```

## 3️⃣ 自定义爬取任务

```python
from ai_smart_crawler import CrawlTask

# 创建自定义任务
task = CrawlTask(
    url="https://example.com",
    timeout=30,
    headers={"User-Agent": "MyBot/1.0"}
)

result = await crawler.crawler.crawl(task)
print(f"状态: {result.success}")
print(f"标题: {result.metadata.get('title', 'N/A')}")
print(f"内容长度: {len(result.markdown)}")
```

## 4️⃣ 保存结果

```python
import json

# 保存到文件
with open('crawl_result.json', 'w', encoding='utf-8') as f:
    json.dump({
        'url': url,
        'content': content[:1000],  # 只保存前1000字符
        'length': len(content)
    }, f, ensure_ascii=False, indent=2)

print("💾 结果已保存到 crawl_result.json")
```
